```python
import numpy as np
import torch
import pandas as pd
```

```python
df = pd.read_csv('./iris.csv')
df.head()
```

```python
X=df.drop('target', axis=1).values
y=df['target'].values
```

```python
from torch.utils.data import TensorDataset, DataLoader
help(TensorDataset)
```

```python
iris = TensorDataset(torch.FloatTensor(X), torch.FloatTensor(y))
iris.tensors
```

```python
len(iris)
```

```python
for i in iris:
    print(i)
```

```python
#batches we use DataLoader
iris_loader = DataLoader(iris,batch_size=50,shuffle=True)

for batch_no, batch in enumerate(iris_loader):
    print(batch_no,'    ', batch)
```
